### IndusInd Bank

```python
df[(df['Bank_Name']=='IndusInd Bank')&(df['Bank_Rating']==5.0)]['Bank_Type'].value_counts()
```

```python
df[(df['Bank_Name']=='IndusInd Bank')&(df['Bank_Rating']==5.0)]['Bank_Type'].value_counts().plot.bar()
plt.title('IndusInd Bank 5.0 Rating Bank_Types')
plt.xlabel('Bank_Type')
```

#### Observayions:

- In IdusInd Bank with 5.0 rating of Credit Card Bank_Types are very high

- Personal Loan types are 19 in same bank.
